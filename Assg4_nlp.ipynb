```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = [
    "Arpita loves machine learning",
    "Machine learning is interesting",
    "Natural language processing is fun",
    "Arpita enjoys learning new things",
    "Language processing and machine learning are related"
]
```

```python
tokenized_text = []

for sentence in text:
    tokens = word_tokenize(sentence.lower())
    tokenized_text.append(tokens)

print(tokenized_text)
```

```python
model = Word2Vec(
    sentences=tokenized_text,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4
)
```

```python
print(model.wv['learning'])
```

```python
similar_words = model.wv.most_similar('learning')
print(similar_words)
```

```python
result = model.wv.most_similar(positive=['machine', 'language'], negative=['learning'])
print(result)
```

```python
result = model.wv.most_similar(positive=['interesting', 'language'], negative=['learning'])
print(result)
```
