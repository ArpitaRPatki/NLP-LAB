```python
# Fix the typo: gensim, not genism
!pip install gensim

import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
nltk.download('punkt_tab')
```

```python
# 1. Prepare a small corpus
corpus = [
    "I love watching movies",
    "I enjoy watching films",
    "Movies are great entertainment",
    "Films are fun to watch",
    "The cinema is a great place"
]

```

```python
# 2. Tokenize and prepare data for Word2Vec
tokenized_data = [word_tokenize(sentence.lower()) for sentence in corpus]

```

```python
# 3. Train the model
# size=10: the dimensionality of the word vectors
# window=2: the maximum distance between the current and predicted word
# min_count=1: ignores words with total frequency lower than this
model = Word2Vec(sentences=tokenized_data, vector_size=10, window=2, min_count=1, workers=1)

```

```python
# 4. Investigate semantic similarities
word = "movies"
if word in model.wv:
    similar_words = model.wv.most_similar(word, topn=3)
    print(f"Words similar to '{word}':")
    for sim_word, score in similar_words:
        print(f" - {sim_word}: {score:.4f}")

```

```python
# 5. Accessing vector representation
print(f"\nVector representation of 'movies':\n{model.wv['movies']}")
```
