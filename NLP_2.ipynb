```python
import nltk
from nltk.tokenize import word_tokenize

# Download necessary NLTK data for POS tagging
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

```

```python
# 1. Define a short text dataset
corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Data science is an interesting field of study.",
    "Python programming is highly effective for automation."
]

```

```python
# 2. Perform Tokenization and POS Tagging
def pos_tagging_pipeline(text_list):
    results = []
    for sentence in text_list:
        # Tokenize the sentence
        tokens = word_tokenize(sentence)
        # Apply POS tagging
        tags = nltk.pos_tag(tokens)
        results.append(tags)
    return results

```

```python
# 3. Process and Display Results
tagged_data = pos_tagging_pipeline(corpus)

for i, tagged_sentence in enumerate(tagged_data):
    print(f"\nSentence {i+1} Tags:")
    print(tagged_sentence)
```
