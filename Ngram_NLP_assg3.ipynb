```python
import nltk
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = """
Arpita loves machine learning.
Arpita loves natural language processing.
Machine learning is fun.
Language processing is interesting.
"""
```

```python
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
def build_ngrams(tokens, n):
    ngrams = []
    for i in range(len(tokens) - n + 1):
        ngram = tuple(tokens[i:i+n])
        ngrams.append(ngram)
    return ngrams

bigrams = build_ngrams(tokens, 2)
print(bigrams[:10])
```

```python
model = defaultdict(Counter)

for w1, w2 in bigrams:
    model[w1][w2] += 1
```

```python
def predict_next_word(model, word):
    if word in model:
        next_words = model[word]
        return next_words.most_common(1)[0][0]
    else:
        return "No prediction available"
```

```python
input_word = "arpita"
prediction = predict_next_word(model, input_word)

print("Input Word:", input_word)
print("Predicted Next Word:", prediction)
```
