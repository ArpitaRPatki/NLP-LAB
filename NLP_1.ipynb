```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

```

```python
# Download necessary NLTK datasets
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')

```

```python
# 1. Corpus
corpus = "The engineers are engineering better systems. The cats were running quickly."
```

```python

# 2. Tokenization
tokens = word_tokenize(corpus)
print(f"Tokens: {tokens}")

```

```python
# 3. Stemming (Using Porter Stemmer)
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in tokens]
print(f"Stemming: {stemmed}")

```

```python
# 4. Lemmatization (Using WordNet)
lemmatizer = WordNetLemmatizer()
# Note: For accurate lemmatization, Part-of-Speech tagging is typically required
lemmatized = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
print(f"Lemmatization: {lemmatized}")
```
