```python
import fitz  # PyMuPDF
import spacy
import re
from google.colab import files

# 1. Load the NLP model
nlp = spacy.load("en_core_web_sm")

def extract_citations_from_pdf():
    # Prompt for file upload
    print("Please upload your research paper (PDF):")
    uploaded = files.upload()

    if not uploaded:
        print("No file uploaded.")
        return

    # Get the filename
    fname = list(uploaded.keys())[0]

    # 2. Extract text from PDF
    text = ""
    with fitz.open(stream=uploaded[fname], filetype="pdf") as doc:
        for page in doc:
            text += page.get_text("text") + " "

    # 3. Clean and process text
    # Remove excessive newlines and spaces common in PDF exports
    clean_text = re.sub(r'\s+', ' ', text)

    # Process with SpaCy for sentence segmentation
    # This prevents "et al." from being treated as the end of a sentence
    doc = nlp(clean_text)
    sentences = [sent.text.strip() for sent in doc.sents]

    # 4. Define Citation Regex
    # Matches: [1], [1, 2], [10-15] OR (Smith, 2020), (Doe et al., 2019)
    citation_regex = r"(\[\d+(?:,\s*\d+|-\d+)*\]|\([A-Z][a-z]+(?:\s+et\s+al\.)?,\s+\d{4}\))"

    citation_sentences = [s for s in sentences if re.search(citation_regex, s)]

    # 5. Output results
    print(f"\n--- Found {len(citation_sentences)} sentences with citations ---\n")
    for i, sent in enumerate(citation_sentences, 1):
        print(f"{i}. {sent}\n")

# Run the system
extract_citations_from_pdf()
```
