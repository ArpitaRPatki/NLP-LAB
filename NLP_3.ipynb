```python
import nltk
from collections import defaultdict, Counter
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')

# 1. Prepare and tokenize text
text = "the cat sat on the mat the cat ate the food the cat sat on the rug"
tokens = word_tokenize(text.lower())

# 2. Build the N-gram (Bigram) model
def build_bigram_model(tokens):
    bigram_counts = defaultdict(Counter)
    for i in range(len(tokens) - 1):
        word = tokens[i]
        next_word = tokens[i + 1]
        bigram_counts[word][next_word] += 1
    return bigram_counts

model = build_bigram_model(tokens)

# 3. Predict the most probable next word
def predict_next_word(word, model):
    if word in model:
        # Get the word with the highest frequency
        return model[word].most_common(1)[0][0]
    return None

# Test
input_word = "cat"
prediction = predict_next_word(input_word, model)
print(f"The most probable word after '{input_word}' is: '{prediction}'")

input_word = "rug"
prediction = predict_next_word(input_word, model)
print(f"The most probable word after '{input_word}' is: '{prediction}'")

input_word = "mat"
prediction = predict_next_word(input_word, model)
print(f"The most probable word after '{input_word}' is: '{prediction}'")
```
